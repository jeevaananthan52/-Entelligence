```python
import nltk
from nltk.tokenize import regexp_tokenize
from nltk.stem.snowball import SnowballStemmer # Changed to SnowballStemmer

nltk.download('punkt') # SnowballStemmer might need 'punkt'

text = "programmed programming programmer."
words=regexp_tokenize(text.lower(), r'\w+')
filter_words=[]
stemmer=SnowballStemmer('english') # Initialize SnowballStemmer
for word in words:
  filter_words.append(stemmer.stem(word)) # Use stem() method

print(filter_words)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
text = "The cats were running faster than the dogs."
tokens = word_tokenize(text)
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]

print(f"Original Text: {text}")
print(f"Lemmatized Words: {lemmatized_words}")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
data = {'Employee id': [10, 20, 15, 25, 30],
        'Gender': ['M', 'F', 'F', 'M', 'F'],
        'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice'],
        }
df=pd.DataFrame(data)
cat_columns=df.select_dtypes(include='object').columns.tolist()
encoder=OneHotEncoder()
encoded=encoder.fit_transform(df[cat_columns])
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out(cat_columns))
df_ans=pd.concat([df,encoded_df],axis=1)
df_ans.drop(cat_columns,axis=1,inplace=True)
print(df_ans)
```
